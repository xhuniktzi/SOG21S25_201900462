```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('winequality-red.csv', sep=';') 
```

```python
print(df.head())
print(df.info())
```

```python
df_describe = df.describe()
print(df_describe)
```

```python
corr_matrix = df.corr()

print(corr_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlacin')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='quality', data=df)
plt.title('Distribucin de la variable Quality')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='quality', y='alcohol', data=df)
plt.title('Boxplot de Alcohol vs Quality')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='volatile acidity', y='quality', data=df)
plt.title('Diagrama de dispersin: Volatile Acidity vs Quality')
plt.show()
```

## Conclusiones
1. Correlacin entre el alcohol y la calidad: Generalmente, se observa que a mayor contenido de alcohol, tiende a aumentar la calidad del vino (correlacin positiva).
2. Correlacin entre la acidez voltil y la calidad: Suele ser negativa; vinos con alta acidez voltil suelen tener menor calidad.
3. Distribucin de la calidad: En la mayora de los casos, los vinos se concentran en calidades intermedias (por ejemplo, 5 y 6). Es poco comn encontrar vinos con calidades muy altas o muy bajas en este dataset.
